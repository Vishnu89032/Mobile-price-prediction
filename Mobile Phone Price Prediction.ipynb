# Data Visualization & analysis

```python
sns.pairplot(dataset,hue='Memory')
```

# memory vs price range

```python
sns.pointplot(y="Memory", x="Prize", data=dataset)
```

# Battery power vs Price Range

```python
sns.boxplot(x="Prize", y="Battery_", data=dataset)
```

# No of Phones vs Camera megapixels of front and primary camera

```python
plt.figure(figsize=(10,6))
dataset['Front Camera'].hist(alpha=0.5,color='blue',label='Front camera')
dataset['Rear Camera'].hist(alpha=0.5,color='red',label='Primary camera')
plt.legend()
plt.xlabel('MegaPixels')
```
